# Exploratory Data Analysis – AI4I Predictive Maintenance

This notebook explores the AI4I predictive maintenance dataset to understand class imbalance, feature behavior, and how different sensor readings relate to machine failure.

The objective is to identify meaningful patterns within the data that may help predict failures and inform model design decisions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display

df = pd.read_csv("../data/ai4i2020.csv")

display(df.head())
```

## Dataset Overview

The dataset structure, column types, and summary statistics are reviewed to establish a baseline understanding of the data before deeper analysis.

```python
display(df.shape)
display(df.columns)
display(df.info())
display(df.describe())
```

### Initial Observations

- The dataset contains 10,000 observations.
- There are no major missing values.
- Most features are numeric sensor readings.
- The target variable is `Machine failure` (0 = No, 1 = Yes).
- The dataset appears clean and prepared for analysis.

### Acronyms
- TWF – Tool Wear Failure
- HDF – Heat Dissipation Failure
- PWF – Power Failure
- OSF – Overstrain Failure
- RNF – Random Failure

```python
failure_counts = df["Machine failure"].value_counts()
failure_percent = df["Machine failure"].value_counts(normalize=True) * 100

display(failure_counts)
display(failure_percent)

ax = failure_counts.plot(kind="bar")

plt.title("Machine Failure Distribution")
plt.xlabel("Machine Failure (0 = No, 1 = Yes)")
plt.ylabel("Count")

for i, v in enumerate(failure_counts):
    ax.text(i, v + 5, str(v), ha='center')

plt.show()
```

### Target Distribution Analysis

The dataset is highly imbalanced, with the majority of observations labeled as non-failure (0). Failures represent only a small percentage of the data (~3.5%).

This has important modeling implications:

- Accuracy alone will not be a reliable metric.
- A model predicting all zeros would achieve high accuracy while failing to detect breakdowns.
- Recall and F1-score will be prioritized to reduce missed failures.

## Correlation Analysis

```python
numeric_df = df.select_dtypes(include=["number"])

corr_matrix = numeric_df.corr()
display(corr_matrix)

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

## Feature Distributions by Failure Status

```python
plt.figure()
sns.boxplot(x="Machine failure", y="Tool wear [min]", data=df)
plt.title("Tool Wear vs Machine Failure")
plt.show()

plt.figure()
sns.boxplot(x="Machine failure", y="Torque [Nm]", data=df)
plt.title("Torque vs Machine Failure")
plt.show()

plt.figure()
sns.boxplot(x="Machine failure", y="Rotational speed [rpm]", data=df)
plt.title("Rotational Speed vs Machine Failure")
plt.show()
```

### Feature-Level Insights

- Failed machines show higher torque values on average.
- Tool wear appears higher among failed machines.
- Machines fail more frequently at lower RPM ranges.

These patterns suggest mechanical stress and wear-related features may be useful predictors of machine failure.

## EDA Summary
Key findings from exploratory analysis:

- The dataset is clean, structured, and contains no major data quality issues.
- The target variable is highly imbalanced (~3.5% failure rate).
- Torque and tool wear show clear relationships with failure.
- Accuracy alone would be misleading due to class imbalance.
- Model evaluation should prioritize recall and F1-score to reduce missed failures.
