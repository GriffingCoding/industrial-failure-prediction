# Model Comparison – Logistic Regression vs Random Forest

This notebook compares the baseline Logistic Regression model with the Random Forest model.  
We focus on operational metrics like recall, precision, F1-score, and ROC-AUC, and summarize feature importance insights.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Store metrics manually from previous notebooks
metrics = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest'],
    'Accuracy': [0.82, 0.98],
    'Precision (Failure)': [0.14, 0.89],
    'Recall (Failure)': [0.84, 0.57],
    'F1-Score (Failure)': [0.24, 0.70],
    'ROC-AUC': [0.908, 0.962]
})
```

## Comparison Table

We summarize all key evaluation metrics for each model in a single table.  
This helps to quickly identify strengths and weaknesses of each approach.

```python
# Display comparison table
display(metrics)
```

## Visual Comparison of Metrics

Display bar charts for Recall, Precision, F1-Score, and ROC-AUC to visually compare model performance.  
This emphasizes which model better meets operational goals (catching failures vs avoiding false alarms).

```python
# Set style
sns.set(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Metrics to plot
plot_metrics = ['Recall (Failure)', 'Precision (Failure)', 'F1-Score (Failure)', 'ROC-AUC']

for ax, metric in zip(axes.flatten(), plot_metrics):
    sns.barplot(data=metrics, x='Model', y=metric, ax=ax)
    ax.set_title(metric)
    ax.set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

## Model Comparison Insights

- **Logistic Regression**: higher recall (0.84) but low precision (0.14) → catches most failures but many false alarms.  
- **Random Forest**: higher precision (0.89) but lower recall (0.57) → fewer false alarms, but misses more failures.  
- **ROC-AUC**: Random Forest (0.96) outperforms Logistic Regression (0.91), showing better overall separation.  
- **Feature Importance**: Torque and Tool Wear are consistently top predictors, confirming EDA insights.  

**Operational takeaway**:  
- If minimizing missed failures is the priority → Logistic Regression is preferred.  
- If minimizing false alarms is more important → Random Forest is preferred.
